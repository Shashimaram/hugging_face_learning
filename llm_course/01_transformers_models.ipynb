## General Transformer Architecture

Transformers composed of two blocks:

| Block | Description |
|:------|:------------|
| **Encoder** | The Encoder receives an input and builds a representation of it(Its features), This means the model is optimized to acquire understanding from the input. |
| **Decoder** | The decoder uses the encoder's representation (features) along with other inputs to generate a target sequence. This means that the model is optimized for generating outputs. |


This two blocks performes independently depending on the task:

| Model Type | Best Suited For |
|:-----------|:----------------|
| **Encoder-only models** | Good for tasks that require understandingg of the input. |
| **Decoder-only models** | good for generative tasks such as text generation. |
| **Encoder-decoder models** (sequence-to-sequence) | good for generating tasks that require an input such as translation or summarization. |

## How Transformers Solve Tasks

Most models are variant of an encoder, decoder or an encoder-decoder structure.

### Transformer Models for Language

language models are heart of modern NLP, that are desigined to understand and generate human language by learning the startical patters and relationships between the words and tokens.

language models are trained to predict the probalility of a word given the context of surroundings words.

### Two Main Training Approaches

| Approach | Description |
|:---------|:------------|
| **Masked Language Modeling (MLM)** | used by encoder like BERT, it randomly masks some tokens in the input and let the model to predict the masked key-word based on the initial word and next word. This is bidirectional context() looking the words before and after of the masked word |
| **Causal Language Modeling (CLM)** | used in GPT, this approach only predics the next tokens based on all the previous tokens in the sequence. this uses only the context form left. |

## Types of Language Models

| Type | Example | Description |
|:-----|:--------|:------------|
| **Encoder-only models** | BERT | this are models use a bidirectional approach to understand context from both directions. They are best suited for tasks that require deep understaidn of text such as classification. NER, and question answering. |
| **Decoder-only models** | GPT, Llama | this models process text from left to right and are particularly good at text generating tasks. they can complete sentences, write essays or even generating code based on a prompt |
| **Encoder-decoder models** | T5 | this models combine both approaches using an encoder to understand the input and decoder to generate output. they excell at sequence-to-sequence tasks like translation and summarizaiton. |

## Text Generation

Here's a much simpler way to think about what you wrote:

### Big Picture

GPT-2 reads text and tries to **predict the next word, one step at a time**.

---

### Step-by-Step (Simplified)

#### Step 1: Break Text into Pieces (Tokens)

- Words are split into smaller chunks using **Byte Pair Encoding**
- Example: `"playing"` → `"play" + "ing"`

---

#### Step 2: Turn Tokens into Numbers

- Each token becomes a vector (a list of numbers the model understands)
- Add position info so the model knows order:
  - "I ate pizza" ≠ "pizza ate I"

---

#### Step 3: Pass Through the Model (Decoder Blocks)

- The model looks at the sequence using **self-attention**
- But with a rule: it can only look **backward (left), not forward**

So if the sentence is:
> "I love ice…"

It can only see:
> "I love ice"
> NOT the next word ("cream")

---

#### Step 4: Predict Next Word

- The model outputs probabilities for what comes next:
  - "cream" → 80%
  - "water" → 5%
  - etc.

---

#### Step 5: Learn from Mistakes

- During training:
  - The "correct" next word is known
  - The model compares its guess vs the correct answer
  - Uses **cross-entropy loss** to measure how wrong it was
  - Adjusts itself to improve

---

### One-line Summary

> GPT-2 **reads text, looks only at past words, and learns to predict the next word step-by-step**.

---

### Key Difference from BERT

| Model | Behavior |
|:------|:---------|
| GPT-2 | predicts next word (left → right) |
| BERT | fills in missing words anywhere using `[MASK]` |

## Deep Dive into Text Generation Inference with LLMs

inference is a process of using trained LLM to generate human-like text from a given input prompt. language models use their knowledge to train and formulate responses one word at a time. the model leverages learned probabilotes from billions of parameters to pridict and generate the next token sequence.

## The Two-Phase Inference Process

LLms generate text in two main phases: `prefill` and `decode` phase.

### Prefill Phase

its like cooking stage where all ingredents are processes and made ready. This involves three key steps:

| Step | Description |
|:-----|:------------|
| **Tokenization** | converting the input text into tokens( this are the basic building blocks of model understanding.) |
| **Embedding Conversion** | Transforming thses tokesn in to numerical representaion that capture their meaning. |
| **Initial Processing** | Running these embiddings through the model's |

> This phase is computationally intensive because it needs to process all the input tokens at once.

### Decode Phase

after the prefill phase as processes the input. we move to decode phase. this is where actual text generation happens.

The decode phase involves:

| Step | Description |
|:-----|:------------|
| **Attention Computation** | looking back at all previous tokens to understant the context. |
| **Probability Calculation** | Determine the likelihood of each possible next token. |
| **Token Selection** | choosing the next token based in there probabilities |
| **Continuation Check** | Deciding weather to continue or stop generation |

> This phase is memory-intensive because the model needs to keep track of all the previous generated tokens and their relationships.

---

## Understanding Token Selection

```
Raw Logits -> Temperature Scaling -> Top-k filtering -> Top-p sampling -> Selected Tokens
```

| Step | Description |
|:-----|:------------|
| **1. Raw Logits** | think of these are model's initial gut feelings about each possible next word. |
| **2. Temperature Control** | Like a creative dial - higher settings(>1.0) make choices more random and creative, lower settings (<1.0) make them more focuesd and determenistinc |
| **3. Top-p (Nucleus) Sampling** | instead of considering all the possible words. we only look at the most likely ones that add up to our chosen probability threshold (e.g., top 90%) |
| **4. Top-K Filtering** | An alternative approach where we only consider the k most likely results |

## Managing Repetition

one common challenge with llms is that there tendency to repeat themselves to address this we use TWO types:

| Penalty Type | Description |
|:-------------|:------------|
| **Presence Penalty** | A fixed penality applied to any token that has appeared before regardless of how often. this helps the model to prevent the model from reusing them same words |
| **Frequency Penalty** | A scaling penalty that increases based on how often the token has been used. the more the word appears the less likely it is to be chosen again. |
