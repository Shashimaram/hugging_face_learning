##### Instruct Model (SmolLM3-3B): Fine-tuned to follow instructions and engage in conversations. If you ask “What’s the weather like?”, it understands this as a question requiring a response as a new message

### Supervised Fine-Tunning with SmolLM3

##### SFT is a cornerstone of instructiion tuning, its how we transform base language model in an instruction-following assistant.
- sft is a process of continuing to traina pretrained model on task specifif datasets with labeled examples.
- pre-training teaches the mdoel general language understanding
- supervised fine-tuning teaches specific skills and behaviors (like learning to do a specfit task)



Here we are not teaching the model a new knowledge from scratch, instead we are shaping how the existing knowledge is applied.
this approach is efffective because it leverages the rich representatins learned during pre-trained while requring significantly less computational resources from scratch.

SmolLM3-3B-Base trani trained on 11T tokens of general data



Why SFT Works: The Science Behind it:

> SFT is effective vecause it leverages the rich representatins learned during the pre-training while adapting the models behavior patters,
> During SFT the models parameters are fine-tunned through gradient descent on task-specfic examples causing subtle but important changes on how the model process and generates the text.

#### Behavioral Adaptation: The model learns and recognixe instruction patterns and respond appropriately. this involves updating the attention mechanisms to focus on instructions cues in languages and adjusting the output distribution to favor the desired responses. but just by updating models surface layers behavior rather than its underlying knowledge.


#### Task Specilization : Rather tahn entirely new concepts the model learns to apply its existing knowledge in specfic contexts. this is why SFT is much more effective that pre-training. its like refining th exisiting knowledge comes from pre-training while sft teache the model how to format and present this knowledge appropriately.


#### Safety ALignment: Through exposure to carefully curated examples. the model learns to be more helpfull, harmless and honest. this involves both learning what ot say, and what not to say.

SFT is needed when your use case require behavior that differs significantly from general -purpose conversation.

### Traning and configuration:

- Choosing the right hyperparameters is crucial for SFT. the goal is to find the sweet spot where the model learns effectively without overfitting or becoming unstable.

Key Hyperparameters:

> Learning Rate: (5e-5 to 1e-4): Controls how much the model weight canhge with each update.
 - start with 5e-5 for SmolLM3; this is conservative and stable.
 - TOO High: The model becomes unstable; loss oscillates or explodes.
 - Too Low: The model learns very slowly and may not converge in reasonable time.

> Batch Size:(4-16): Numbe of examples processed simultaneously.
 -  larger batches: More stable gradients, but require more GPU meory.
 - smaller batches: less memory usage, but noisier gradients.
 - use grandient accumulation to achienve large effective batch sizes.

> maz sequence length (2048-4096) : Maximul tokens per traninig example:
- longer sequences can handel more complex conversations
- shorter sequences : faster training, less memory usage
- match your case: use the typical length of yout targets conversations
> tranining steps: (1000 - 5000) : total number of parameters to updates
- depends on dataset size: more data usually requires more steps
- monitor validation loss: stop whrn it stops improving
- thumb rule: three to five epochs through your dataset.
> Warmup Steps: (10% of  gradual learning rate increase at start):
- prevents early instability: helps th emdoel to adapt gradually
- typical range: 100-500 ste[s for most sft tasks]

### Monitoring and Evaluation:

Effective monitoring is crucial. unlike pre=training where you primarily wathc loss decrease,SFT requires carefull attention to both quantitative metrics and qualitative outputs. the goal os to ensure your model is learning the desired behaviors.

Key Metrics to Monitor:

Training Loss: Should decrease steadily but not too rapidly

Healthy pattern: Smooth, gradual decrease.
Warning signs: Sudden spikes, oscillations, or plateaus.
Typical range: Starts around 2-4, should decrease to 0.5-1.5.
Validation Loss: Most important metric for preventing overfitting

Should track training loss: A small gap indicates good generalization.
Growing gap: Sign of overfitting; the model may be memorizing training data.
Use for early stopping: Stop training when validation loss stops improving.
Sample Outputs: Regular qualitative checks are essential

Generate responses: Test the model on held-out prompts during training.
Check format consistency: Ensure the model follows desired response patterns.
Monitor for degradation: Watch for repetitive or nonsensical outputs.
Resource Usage: Track GPU memory and training speed

Memory spikes: May indicate batch size is too large.
Slow training: Could suggest inefficient data loading or processing.
